```python
#dataframe maken voor two-way anova
df1 = pd.DataFrame()
df1["treatment"] = ["control"]*9 + ["fmlp"] *9 + ["cxcl1"] * 9 + ["cxcl8"] *9
df1["status"] = (["healthy"] * 3 + ["mild"] * 3 + ["severe"] * 3)*4
df1["values"] = np.concatenate((tabel1["mean_cell_speed"], tabel2["mean_cell_speed"], tabel3["mean_cell_speed"], tabel4["mean_cell_speed"]))
df1
```

```python
#two way anova
import statsmodels.api as sm
from statsmodels.formula.api import ols

mod = ols('values ~ treatment + status',
data=df1).fit()
aov_table = sm.stats.anova_lm(mod, typ=2)
print (aov_table)
```

```python
#two way anova with interaction
mod = ols('values ~ treatment * status', #nu * ipv +
data=df1).fit()
aov_table = sm.stats.anova_lm(mod, typ=2)
print (aov_table)
```

Als men voor alfa een waarde 0.05 vooropstelt dan is er een significante interactie tussen de twee factoren treatment en status, want de p-waarde = 0.0353. (Niet als alfa = 0.01) (welke kies je dan?)
De p-waarden van de factoren zijn nu ook verschillend tov die erboven zonder interactie, maar nog steeds laag (lager zelfs)
Dus er is wel degelijk een verschil tussen de gemiddelde waardes van de snelheid bij verschillende treatment en verschillende status???
