### 1. Accumulated distance

De totale afstand die de cellen afleggen: som van alle delta'z

```python
accumulated_distance_list = []
accumulated_distance = 0.00
for patient in tracks_final.patientnumber.unique():
    print('current patient: {}'.format(patient))
    tmp3 = tracks_final[tracks_final.patientnumber==patient]
    for cellType_treatment in tmp3.cellType_treatment.unique():
        print('current condition: {}'.format(cellType_treatment))
        tmp = tmp3[tmp3.cellType_treatment==cellType_treatment]

        for track in tmp.track_id.unique():
            tmp2 = tmp[tmp.track_id==track]
            array = tmp2.drop(['track_id', 't', 'cellType_treatment', 'norm_x', 'norm_y', 'x', 'y', 'patientnumber', 'status', 'delta_x', 'delta_y', 'ta_deg'], axis=1).as_matrix()
            for element in tmp2["delta_z"]:
                if math.isnan(element):
                    pass
                else:
                    accumulated_distance += element
           
            accumulated_distance_list.append(accumulated_distance)
            accumulated_distance = 0.00
```

### 2. Accumulated euclidian distance

De afstand tussen begin en eindpunt van een track

(Cellen kunnen op hun stappen terugkeren, wat de euclidische afstand verkleind, maar de geaccumuleerde afstand vergroot!)

Euclidian distance tussen P(x1, y1) en Q(x2, y2) = sqrt((y1 - x1)^2 + (y2-x2)^2)

```python
accumulated_euclidian_distance_list = []
for patient in tracks_final.patientnumber.unique():
    print('current patient: {}'.format(patient))
    tmp3 = tracks_final[tracks_final.patientnumber==patient]
    for cellType_treatment in tmp3.cellType_treatment.unique():
        print('current condition: {}'.format(cellType_treatment))
        tmp = tmp3[tmp3.cellType_treatment==cellType_treatment]

        for track in tmp.track_id.unique():
            tmp2 = tmp[tmp.track_id==track]
            #array met enkel nog x en y erin:
            array = tmp2.drop(['track_id', 't', 'cellType_treatment', 'norm_x', 'norm_y', 'patientnumber', 'status', 'delta_x', 'delta_y', 'delta_z', 'ta_deg'], axis=1).as_matrix()

            accumulated_euclidian_distance = math.sqrt((array[0,1] - array[0,0])**2 + (array[-1,1] - array[-1, 0])**2)
           
            accumulated_euclidian_distance_list.append(accumulated_euclidian_distance)
            
```

```python
print(accumulated_euclidian_distance_list[:100])
len(accumulated_euclidian_distance_list)
#handmatig uitgerekend en het klopt
```

### 3. Average directness (directionality)

De verhouding tussen de accumulated euclidian distance en de accumulated distance

It is a measure of the straightness of the cell trajectories. A directionality of D --> 1 equals a straight-line migration from start to endpoint.

```python
average_directness = 0
average_directness_list = []
for i in range(9792):
    if accumulated_distance_list[i] != 0:
        average_directness = accumulated_euclidian_distance_list[i]/accumulated_distance_list[i]
    else:
        average_directness = "NaN"
    average_directness_list.append(average_directness)
```

### 4. Speed

Accumulated distance / laatste tijdspunt  (dus gemiddelde snelheid per track)

Elk tijdspunt staat voor 8 seconden!

Voor elke track moet je laatste tijdspunt verkijgen en dat dan maal 8

```python
tijdspunten_lijst = []
for patient in tracks_final.patientnumber.unique():
    print('current patient: {}'.format(patient))
    tmp3 = tracks_final[tracks_final.patientnumber==patient]
    for cellType_treatment in tmp3.cellType_treatment.unique():
        print('current condition: {}'.format(cellType_treatment))
        tmp = tmp3[tmp3.cellType_treatment==cellType_treatment]

        for track in tmp.track_id.unique():
            tmp2 = tmp[tmp.track_id==track]
            array = tmp2.drop(['track_id', 'cellType_treatment', 'norm_x', 'norm_y', 'patientnumber', 'status', 'delta_x', 'delta_y', 'delta_z', 'ta_deg', 'x', 'y'], axis=1).as_matrix()
            tijdspunt = int(max(array))
            tijdspunten_lijst.append(tijdspunt)
           
```

```python
print(len(tijdspunten_lijst))
tijdspunten_lijst[:5]
```

```python
#In seconden nu:
tijdspunten_lijst_in_sec = [i * 8 for i in tijdspunten_lijst]
tijdspunten_lijst_in_sec[:5]
```

```python
speed_list = []
for i in range(9792):
    speed = accumulated_distance_list[i]/tijdspunten_lijst_in_sec[i]
    speed_list.append(speed)
```

```python
print(len(speed_list))
speed_list[:5]
```

### 5. Velocity - nu stdev

De afgeleide van de snelheid naar de tijd

```python
#velocity per step: (al in seconden): dx/dt (het verschil in tijd tussen twee stappen is altijd 8 seconden)
velocity_list = []
tussentijdse = []
for patient in tracks_final.patientnumber.unique():
    print('current patient: {}'.format(patient))
    tmp3 = tracks_final[tracks_final.patientnumber==patient]
    for cellType_treatment in tmp3.cellType_treatment.unique():
        print('current condition: {}'.format(cellType_treatment))
        tmp = tmp3[tmp3.cellType_treatment==cellType_treatment]

        for track in tmp.track_id.unique():
            tmp2 = tmp[tmp.track_id==track]
            array = tmp2.drop(['track_id', 't', 'cellType_treatment', 'norm_x', 'norm_y', 'patientnumber', 'status', 'delta_x', 'delta_y', 'ta_deg', 'x', 'y'], axis=1).as_matrix()
            for element in array:  #array met delta-z er nog in
                if math.isnan(element):
                    velocity = 0.0
                else:
                    velocity = float(element) / 8.00
                tussentijdse.append(velocity)

            #average velocity per track:
            velocity_track = np.std(tussentijdse)
            velocity_list.append(velocity_track)
            tussentijdse = []
```

```python
print(velocity_list[:20])
len(velocity_list)
#als het mean is die gebruikt wordt is jet hetzelfde als de snelheid
# moet ik dat nu nog op een andere manier doen, of is het dit wat ze in de vorige data hadden gebruikt?
```
