```python
from scipy import stats
```

```python
# 2 sample t-test
stats.ttest_ind(tabel1["mean_cell_speed"], tabel2["mean_cell_speed"])
```

```python
#paired test
stats.ttest_rel(tabel1["mean_cell_speed"], tabel2["mean_cell_speed"])
```

```python
#paired non parametric test
stats.wilcoxon(tabel1["mean_cell_speed"], tabel2["mean_cell_speed"])
```

```python
#ANOVA
stats.f_oneway(tabel1["mean_cell_speed"], tabel2["mean_cell_speed"], tabel3["mean_cell_speed"], tabel4["mean_cell_speed"])
```

```python
stats.wilcoxon(tabel1["mean_cell_speed"], tabel3["mean_cell_speed"])
```
